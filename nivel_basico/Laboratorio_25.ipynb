```python
import datetime

# Datos simulados de disponibilidad de mesas
availability = {
     "2024-10-10": ["18:00", "20:00"],
     "2024-10-11": ["17:00", "19:00", "21:00"],
     "2024-10-12": ["18:00", "20:00"]
     }
# Diccionario para guardar reservas
reservations = {}
```

```python
# Función para obtener la fecha de reserva del usuario
def get_reservation_date():
  print("\n¿Cuál es la fecha de la reserva que te interesa? (Formato: YYYY-MM-DD)")
  date_str = input("> ")
  try:
    date = datetime.datetime.strptime(date_str, "%Y-%m-%d").date()
    if date_str in availability:
      return date_str
    else:
      print("Lo siento, no tenemos disponibilidad para esa fecha.")
      return None
  except ValueError:
      print("Formato de fecha inválido. Intenta nuevamente.")
      return None
```

```python
# Función para mostrar la disponibilidad de horarios en una fecha específica
def check_availability():
  date_str = get_reservation_date()
  if date_str:
    times = availability[date_str]
    print(f"\nLos horarios disponibles para {date_str} son: {', '.join(times)}")
```

```python
# Función para hacer una reserva
def make_reservation(user_name):
  date_str = get_reservation_date()
  if date_str:
    times = availability[date_str]
    print(f"\nHorarios disponibles para {date_str}: {', '.join(times)}")
    print("¿A qué hora deseas hacer la reserva? (Formato: HH:MM)")
    time = input("> ")
    if time in times:
      reservations[user_name] = {"date": date_str, "time": time}
      print(f"\nReserva confirmada para {user_name} el {date_str} a las {time}.")
      # Remover horario reservado de la disponibilidad
      times.remove(time)
    else:
        print("Ese horario no está disponible. Intenta nuevamente.")
  else:         print("Fecha no válida. Intenta nuevamente.")
```

```python
# Función para consultar el estado de una reserva
def check_reservation(user_name):
  if user_name in reservations:
    reservation = reservations[user_name]
    print(f"\nTu reserva está confirmada para el {reservation['date']} a las    {reservation['time']}.")
  else:
    print("No tienes ninguna reserva confirmada.")
```

```python
# Función principal para ejecutar el asistente virtual
def virtual_assistant():
  print("¡Hola! Soy tu asistente virtual de reservas. ¿Cuál es tu nombre?")
  user_name = input("> ")
  print(f"\nEncantado de ayudarte, {user_name}. ¿En qué puedo asistirte hoy?")

  while True:
    print("\nPor favor, elige una opción:")
    print("1. Consultar disponibilidad de reservas")
    print("2. Hacer una reserva")
    print("3. Consultar mi reserva")
    print("4. Salir")

    option = input("> ")
    if option == "1":
      check_availability()
    elif option == "2":
      make_reservation(user_name)
    elif option == "3":
      check_reservation(user_name)
    elif option == "4":
      print("Gracias por usar el asistente virtual de reservas. ¡Hasta pronto!")
      break
    else:
      print("Opción no válida. Intenta nuevamente.")

# Ejecución del asistente virtual
virtual_assistant()
```
