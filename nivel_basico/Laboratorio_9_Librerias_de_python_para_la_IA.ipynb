```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "Mes":["Enero","Febrero","Marzo","Abril","Mayo","Junio","Julio","Agosto","Septiembre","Octubre","Noviembre","Diciembre"],
    "Ventas":[25,30,28,45,50,63,56,90,36,73,85,12]
    }

df = pd.DataFrame(data)
media = df["Ventas"].mean()
desviacion = df["Ventas"].std()

df["es_mayor_a_la_media"]= df['Ventas'] > media
plt.bar(df["Mes"],df["Ventas"], color="orange")
plt.title("Ventas Mensuales")
plt.xlabel("Meses")
plt.ylabel("Ventas")
plt.xticks(rotation=90)
plt.axhline(media, color="red", linestyle="--", label="Media de ventas")
```
