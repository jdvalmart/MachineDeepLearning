# 🧪 Ejercicio 1: Preprocesamiento de Datos Textuales

En este ejercicio aprenderás a limpiar y preparar texto para análisis de sentimientos dentro del Procesamiento de Lenguaje Natural (NLP).

## 🎯 Objetivos
- Convertir texto a minúsculas
- Eliminar puntuación
- Tokenizar texto
- Eliminar stopwords

```python
# 📦 Importar bibliotecas necesarias
import nltk
import re
from nltk.corpus import stopwords

# Descargar recursos
nltk.download('stopwords')
```

## 📝 Texto de ejemplo

Usaremos un texto que contiene opiniones mixtas para analizarlo.

```python
# Definir texto de ejemplo
texto = "¡El servicio fue excelente, pero el tiempo de espera fue demasiado largo!"
print("Texto original:\n", texto)
```

## 🔡 Convertir a minúsculas

Esto ayuda a evitar duplicados como "Excelente" y "excelente".

```python
# Convertir a minúsculas
texto = texto.lower()
print("Texto en minúsculas:\n", texto)
```

## 🧹 Eliminar puntuación y caracteres especiales

Usamos expresiones regulares para limpiar el texto.

```python
# Eliminar puntuación
texto = re.sub(r'[^\w\s]', '', texto)
print("Texto limpio:\n", texto)
```

## ✂️ Tokenización

Dividimos el texto en palabras individuales (tokens).

```python
# Tokenización
palabras = texto.split()
print("Tokens:\n", palabras)
```

## 🚫 Eliminación de stopwords

Quitamos palabras comunes que no aportan significado.

```python
# Eliminar stopwords
stop_words = set(stopwords.words('spanish'))

texto_filtrado = [palabra for palabra in palabras if palabra not in stop_words]

print("Texto preprocesado:\n", texto_filtrado)
```

## ✅ Resultado final

El texto ha sido:
- Limpiado
- Tokenizado
- Filtrado

Ahora está listo para ser usado en modelos de Machine Learning.

# 🧠 Ejercicio 2: Análisis de Sentimientos con TextBlob

**Tiempo estimado:** 25 minutos  

## 🎯 Objetivo
Realizar un análisis de sentimientos basado en léxicos utilizando la librería `TextBlob`, identificando si una frase es positiva, negativa o neutral a partir de su polaridad.

## 📖 Descripción

En este ejercicio utilizaremos la librería `TextBlob` para analizar el sentimiento de varias frases.

TextBlob calcula un valor llamado **polaridad**, que indica:

- **-1** → Sentimiento muy negativo  
- **0** → Sentimiento neutral  
- **+1** → Sentimiento muy positivo  

A partir de este valor, clasificaremos cada frase como:
- Positiva
- Negativa
- Neutral

```python
from textblob import TextBlob
```

## 📝 Frases de ejemplo

A continuación, definimos un conjunto de frases para analizar su sentimiento.

```python
frases = [
    "El producto es increíble y super útil.",
    "Muy mala calidad, no lo recomiendo.",
    "La entrega fue rápida, pero el empaque estaba dañado."
]
```

## 🔍 Análisis de sentimientos

Para cada frase:
1. Se calcula la polaridad usando TextBlob
2. Se clasifica el sentimiento según el valor obtenido

```python
for frase in frases:
    blob = TextBlob(frase)
    polaridad = blob.sentiment.polarity

    if polaridad > 0:
        sentimiento = "Positivo"
    elif polaridad < 0:
        sentimiento = "Negativo"
    else:
        sentimiento = "Neutral"

    print(f"Frase: '{frase}' | Polaridad: {polaridad:.2f} | Sentimiento: {sentimiento}")
```

## 📊 Conclusiones

- TextBlob permite realizar análisis de sentimientos de forma sencilla.
- La polaridad es una métrica útil para clasificar emociones en texto.
- Este enfoque es rápido, pero tiene limitaciones, especialmente en español.
- Para mejores resultados, se pueden usar modelos más avanzados como BERT o modelos entrenados específicamente para español.

# 🧠 Ejercicio 3: Análisis de Sentimientos con Naive Bayes

**Tiempo estimado:** 30 minutos  

## 🎯 Objetivo
Implementar un modelo de clasificación de texto utilizando Naive Bayes y TF-IDF para realizar análisis de sentimientos.

## 📖 Descripción

En este ejercicio construiremos un modelo de Machine Learning para clasificar sentimientos en texto.

Se utilizarán las siguientes técnicas:
- **TF-IDF (Term Frequency - Inverse Document Frequency)** para convertir texto en números
- **Naive Bayes** como modelo de clasificación

El flujo será:
1. Vectorizar el texto
2. Dividir los datos en entrenamiento y prueba
3. Entrenar el modelo
4. Evaluar su rendimiento

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

## 📝 Datos de entrenamiento

Definimos un conjunto simple de frases con sus respectivas etiquetas:

- **1 → Positivo**
- **0 → Negativo**

```python
frases = [
    "El producto es excelente",
    "Muy mala calidad",
    "Totalmente satisfecho",
    "No lo recomiendo",
    "Buena relación calidad-precio",
    "Terrible experiencia",
    "Muy contento con la compra",
    "Definitivamente no es lo que esperaba"
]

etiquetas = [1, 0, 1, 0, 1, 0, 1, 0]
```

## 🔢 Vectorización con TF-IDF

Convertimos las frases en vectores numéricos utilizando TF-IDF.

Esto permite que el modelo entienda el texto como datos matemáticos.

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(frases)
```

## 🔀 División de datos

Dividimos el dataset en:
- **Entrenamiento (75%)**
- **Prueba (25%)**

Esto nos permite evaluar el rendimiento del modelo.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, etiquetas, test_size=0.25, random_state=42
)
```

## 🤖 Entrenamiento del modelo

Utilizamos el algoritmo **Multinomial Naive Bayes**, muy común en clasificación de texto.

```python
modelo = MultinomialNB()
modelo.fit(X_train, y_train)
```

## 📊 Evaluación del modelo

Evaluamos el modelo utilizando la métrica de **precisión (accuracy)**.

```python
predicciones = modelo.predict(X_test)
precision = accuracy_score(y_test, predicciones)

print("Precisión del modelo:", precision)
```

```python
nuevas_frases = ["Me encantó el producto", "Es una terrible compra"]
X_nuevo = vectorizer.transform(nuevas_frases)
preds = modelo.predict(X_nuevo)

for frase, pred in zip(nuevas_frases, preds):
    print(frase, "->", "Positivo" if pred == 1 else "Negativo")
```

```python
print(vectorizer.get_feature_names_out())
```

## 📊 Conclusiones

- Naive Bayes es un modelo simple pero efectivo para clasificación de texto.
- TF-IDF mejora la representación del texto frente a métodos básicos.
- El dataset es muy pequeño, por lo que la precisión puede no ser representativa.
- Para mejorar:
  - Usar más datos
  - Aplicar limpieza de texto (stopwords, lematización)
  - Probar modelos más avanzados (Logistic Regression, SVM, BERT)

# 🧠 Ejercicio 4: Evaluación del Modelo y Análisis de Sesgos

**Tiempo estimado:** 20 minutos  

## 🎯 Objetivo
Evaluar el rendimiento del modelo de Naive Bayes utilizando métricas como matriz de confusión, precisión, recall y F1-score, e identificar posibles sesgos.

## 📖 Descripción

En este ejercicio analizaremos el desempeño del modelo entrenado previamente.

Utilizaremos:
- **Matriz de confusión**
- **Precisión (Precision)**
- **Recall**
- **F1-Score**

Estas métricas permiten evaluar no solo qué tan preciso es el modelo, sino también cómo se comporta con cada clase.

```python
from sklearn.metrics import confusion_matrix, classification_report
```

## 🔍 Matriz de Confusión

La matriz de confusión muestra:

- Verdaderos positivos (TP)
- Verdaderos negativos (TN)
- Falsos positivos (FP)
- Falsos negativos (FN)

Esto permite entender en qué tipo de errores está fallando el modelo.

```python
matriz_confusion = confusion_matrix(y_test, predicciones)
print("Matriz de Confusión:\n", matriz_confusion)
```

## 📊 Reporte de Clasificación

El reporte incluye:

- **Precisión**: Qué tan correctas son las predicciones positivas  
- **Recall**: Qué tantos positivos reales detecta el modelo  
- **F1-Score**: Balance entre precisión y recall  

Estas métricas se calculan por cada clase (Negativo y Positivo).

```python
reporte = classification_report(
    y_test,
    predicciones,
    target_names=["Negativo", "Positivo"]
)

print("Reporte de Clasificación:\n", reporte)
```

## 🧠 Interpretación de Resultados

- Si la **precisión es alta**, el modelo comete pocos falsos positivos.
- Si el **recall es alto**, el modelo detecta la mayoría de los casos positivos.
- El **F1-score** es útil cuando hay desbalance entre clases.

### 🔎 Observaciones posibles:
- El modelo puede predecir más una clase que otra.
- Puede fallar en detectar correctamente frases negativas o positivas.
