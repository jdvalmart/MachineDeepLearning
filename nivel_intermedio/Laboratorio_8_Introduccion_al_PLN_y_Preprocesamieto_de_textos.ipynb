# 🧪 Laboratorio 8: Preprocesamiento y Tokenización

En este ejercicio aprenderás a limpiar y preprocesar texto para prepararlo para modelos de Machine Learning dentro del campo del Procesamiento de Lenguaje Natural (NLP).

## 🎯 Objetivos
- Limpiar texto
- Aplicar tokenización
- Eliminar stopwords
- Preparar datos para modelos

```python
# 📦 Importar bibliotecas
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import re

# Descargar recursos necesarios
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

## 📊 Cargar los datos

Crearemos un conjunto de datos de ejemplo con comentarios clasificados como positivos o negativos.

```python
# Crear DataFrame con datos de ejemplo
data = {
    'texto': [
        "Me encanta este producto, es maravilloso!",
        "El servicio fue terrible, no lo recomendaría.",
        "La calidad es excelente, me gusta mucho.",
        "No vale la pena, muy mala experiencia."
    ],
    'etiqueta': ["positivo", "negativo", "positivo", "negativo"]
}

df = pd.DataFrame(data)
df
```

## 🧹 Preprocesamiento del texto

Realizaremos los siguientes pasos:
- Convertir a minúsculas
- Eliminar caracteres especiales
- Tokenizar el texto
- Eliminar stopwords

```python
# Definir stopwords en español
stop_words = set(stopwords.words('spanish'))

def preprocesar_texto(texto):
    # 1. Convertir a minúsculas
    texto = texto.lower()
    
    # 2. Eliminar caracteres especiales
    texto = re.sub(r'[^a-zñáéíóú ]', '', texto)
    
    # 3. Tokenización
    palabras = word_tokenize(texto)
    
    # 4. Eliminar stopwords
    palabras = [palabra for palabra in palabras if palabra not in stop_words]
    
    # 5. Unir palabras nuevamente
    return ' '.join(palabras)
```

## ⚙️ Aplicar preprocesamiento

Aplicamos la función a cada fila del dataset.

```python
# Aplicar preprocesamiento
df['texto_procesado'] = df['texto'].apply(preprocesar_texto)

df[['texto', 'texto_procesado']]
```

## ✅ Resultados

Observamos cómo el texto ha sido limpiado:

- Se eliminaron palabras irrelevantes
- Se redujo el ruido
- El texto está listo para modelos de Machine Learning

# 🔢 Ejercicio 2: Representación del Texto

En este ejercicio convertiremos texto en datos numéricos usando:

- Bag of Words (BoW)
- TF-IDF

Esto es necesario para entrenar modelos de Machine Learning.

## 🧠 ¿Por qué necesitamos esto?

Los modelos de Machine Learning no entienden texto, solo números.

Por eso transformamos:

"encanta producto"

en vectores numéricos.

## 📦 Bag of Words (BoW)

Cuenta cuántas veces aparece cada palabra en el texto.

```python
from sklearn.feature_extraction.text import CountVectorizer

# Crear el vectorizador BoW
vectorizer = CountVectorizer()

# Transformar el texto procesado
X_bow = vectorizer.fit_transform(df['texto_procesado'])

# Mostrar resultados
print("📊 Matriz de Características (BoW):\n")
print(X_bow.toarray())

print("\n🧾 Características (palabras):\n")
print(vectorizer.get_feature_names_out())
```

## 🔍 Interpretación de BoW

- Cada fila = un comentario
- Cada columna = una palabra
- El valor = número de veces que aparece la palabra

## ⚖️ TF-IDF

TF-IDF mide la importancia de una palabra:

- Penaliza palabras muy comunes
- Resalta palabras más relevantes

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Crear el vectorizador TF-IDF
tfidf_vectorizer = TfidfVectorizer()

# Transformar el texto procesado
X_tfidf = tfidf_vectorizer.fit_transform(df['texto_procesado'])

# Mostrar resultados
print("📊 Matriz de Características (TF-IDF):\n")
print(X_tfidf.toarray())

print("\n🧾 Características (palabras):\n")
print(tfidf_vectorizer.get_feature_names_out())
```

## 🔍 Interpretación de TF-IDF

- Valores más altos = palabras más importantes
- Valores cercanos a 0 = palabras menos relevantes

A diferencia de BoW, aquí no solo contamos, sino que ponderamos.

## ⚔️ BoW vs TF-IDF

| Método | Ventaja | Desventaja |
|--------|--------|-----------|
| BoW | Simple y rápido | No mide importancia |
| TF-IDF | Más inteligente | Más costoso |


# 🤖 Ejercicio 3: Clasificador de Texto con Naive Bayes

En este ejercicio entrenaremos un modelo de Machine Learning para clasificar textos como positivos o negativos.

## 🎯 Objetivos
- Dividir datos en entrenamiento y prueba
- Entrenar un modelo Naive Bayes
- Evaluar el rendimiento

## ✂️ División de los datos

Dividimos el dataset en:

- Entrenamiento (70%)
- Prueba (30%)

Esto permite evaluar el modelo con datos que no ha visto.

```python
from sklearn.model_selection import train_test_split

# Dividir los datos
X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, 
    df['etiqueta'], 
    test_size=0.3, 
    random_state=42
)

print("Tamaño entrenamiento:", X_train.shape)
print("Tamaño prueba:", X_test.shape)
```

## 🧠 Modelo Naive Bayes

Naive Bayes es un algoritmo probabilístico basado en el teorema de Bayes.

Es muy usado en:
- Clasificación de texto
- Filtros de spam
- Análisis de sentimientos

```python
from sklearn.naive_bayes import MultinomialNB

# Crear el modelo
clasificador = MultinomialNB()

# Entrenar el modelo
clasificador.fit(X_train, y_train)
```

## 🔮 Predicciones

Usamos el modelo entrenado para predecir nuevas etiquetas.

```python
# Realizar predicciones
y_pred = clasificador.predict(X_test)

print("Predicciones:", y_pred)
print("Valores reales:", y_test.values)
```

## 📊 Evaluación del modelo

Evaluamos el rendimiento usando:

- Accuracy (precisión)
- Classification Report

```python
from sklearn.metrics import accuracy_score, classification_report

# Precisión
print("✅ Precisión del modelo:", accuracy_score(y_test, y_pred))

# Reporte completo
print("\n📊 Reporte de clasificación:\n")
print(classification_report(y_test, y_pred))
```

## 🔍 Interpretación

- Accuracy cercano a 1 → modelo bueno
- Precision → qué tan exacto es
- Recall → qué tanto detecta correctamente
- F1-score → balance entre precision y recall

⚠️ Nota: Con pocos datos, el modelo puede no ser muy preciso.
