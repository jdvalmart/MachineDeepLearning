# 📊 Laboratorio: Bases de Datos con SQLite y Python

En este laboratorio se implementa un sistema básico de gestión académica utilizando SQLite y Python.

## 🎯 Objetivos
- Crear y estructurar bases de datos relacionales
- Insertar, consultar, actualizar y eliminar datos
- Aplicar consultas avanzadas (JOIN, filtros, ordenamiento)

## 🛠 Tecnologías utilizadas
- Python 3
- SQLite (sqlite3)

```python
import sqlite3

# Conexión a la base de datos (se crea si no existe)
conexion = sqlite3.connect('laboratorio.db')
cursor = conexion.cursor()

print("✅ Conexión establecida")
```

## 🏗 Creación de Tablas

Se crean las siguientes tablas:
- Estudiantes
- Cursos
- Inscripciones (relación entre ambas)

```python
# Tabla Estudiantes
cursor.execute("""
CREATE TABLE IF NOT EXISTS Estudiantes (
    ID INTEGER PRIMARY KEY,
    Nombre TEXT NOT NULL,
    Edad INTEGER NOT NULL,
    Carrera TEXT NOT NULL
)
""")

# Tabla Cursos
cursor.execute("""
CREATE TABLE IF NOT EXISTS Cursos (
    CursoID INTEGER PRIMARY KEY,
    Nombre TEXT NOT NULL,
    Creditos INTEGER NOT NULL
)
""")

# Tabla Inscripciones
cursor.execute("""
CREATE TABLE IF NOT EXISTS Inscripciones (
    InscripcionID INTEGER PRIMARY KEY,
    EstudianteID INTEGER,
    CursoID INTEGER,
    FOREIGN KEY (EstudianteID) REFERENCES Estudiantes(ID),
    FOREIGN KEY (CursoID) REFERENCES Cursos(CursoID)
)
""")

conexion.commit()
print("✅ Tablas creadas correctamente")
```

## 📥 Inserción de Datos

Se insertan registros en las tablas utilizando `executemany` para eficiencia.

```python
# Insertar estudiantes
cursor.executemany("""
INSERT OR IGNORE INTO Estudiantes (ID, Nombre, Edad, Carrera)
VALUES (?, ?, ?, ?)
""", [
    (1, "Laura Gómez", 22, "Ingeniería"),
    (2, "Juan Pérez", 24, "Matemáticas"),
    (3, "Ana Torres", 21, "Física")
])

# Insertar cursos
cursor.executemany("""
INSERT OR IGNORE INTO Cursos (CursoID, Nombre, Creditos)
VALUES (?, ?, ?)
""", [
    (101, "Cálculo", 4),
    (102, "Programación", 3),
    (103, "Física General", 4)
])

# Insertar inscripciones
cursor.executemany("""
INSERT OR IGNORE INTO Inscripciones (InscripcionID, EstudianteID, CursoID)
VALUES (?, ?, ?)
""", [
    (1, 1, 101),
    (2, 2, 102),
    (3, 3, 103)
])

conexion.commit()
print("✅ Datos insertados correctamente")
```

## 🔍 Consultas Básicas

Visualización de los datos almacenados en las tablas.

```python
print("📌 Estudiantes:")
cursor.execute("SELECT * FROM Estudiantes")
for fila in cursor.fetchall():
    print(fila)

print("\n📌 Cursos:")
cursor.execute("SELECT * FROM Cursos")
for fila in cursor.fetchall():
    print(fila)
```

## ✏️ Actualización de Datos

Se actualiza la carrera de un estudiante.

```python
cursor.execute("""
UPDATE Estudiantes
SET Carrera = 'Ciencias de Datos'
WHERE ID = 1
""")

conexion.commit()
# 🔍 Consultar el registro actualizado
cursor.execute("SELECT * FROM Estudiantes WHERE ID = 1")
estudiante = cursor.fetchone()

print("✅ Registro actualizado:")
print(estudiante)
```

## ❌ Eliminación de Datos

Se elimina un curso específico.

```python
cursor.execute("""
DELETE FROM Cursos
WHERE CursoID = 102
""")

conexion.commit()
print("\n📌 Cursos:")
cursor.execute("SELECT * FROM Cursos")
for fila in cursor.fetchall():
    print(fila)

print("✅ Curso eliminado")
```

## 📊 Consulta Avanzada

Filtrar estudiantes mayores de 21 años y ordenarlos por nombre.

```python
cursor.execute("""
SELECT * FROM Estudiantes
WHERE Edad > 21
ORDER BY Nombre ASC
""")

for fila in cursor.fetchall():
    print(fila)
```

## 🔗 Consulta con JOIN

Se combinan las tablas para obtener información relacional.

```python
cursor.execute("""
SELECT Estudiantes.Nombre, Cursos.Nombre, Cursos.Creditos
FROM Inscripciones
JOIN Estudiantes ON Inscripciones.EstudianteID = Estudiantes.ID
JOIN Cursos ON Inscripciones.CursoID = Cursos.CursoID
""")

for fila in cursor.fetchall():
    print(f"Estudiante: {fila[0]}, Curso: {fila[1]}, Créditos: {fila[2]}")
```

## 🔒 Cierre de Conexión

Se cierra la conexión para liberar recursos.

```python
conexion.close()
print("✅ Conexión cerrada")
```
